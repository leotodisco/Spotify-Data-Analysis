```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("../data/top1000_songs.csv")

df
```

```python
df.describe()
```

Count the songs for each decade

```python
df.groupby("decade").count()
```

Average tempo per decade

```python
tempo_df = df.groupby("decade").tempo.agg(["min", "max", "mean"])

tempo_df
```

Boxplot For Tempo column across different decades

```python
plt.figure(figsize=(12, 6))
df.boxplot(column='tempo', by='decade')
plt.title('Distribuzione dei BPM (Tempo) per Decade')
plt.suptitle('')
plt.xlabel('Decade')
plt.ylabel('Tempo (BPM)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
mean_tempo = df.groupby("decade")["tempo"].mean().reset_index()
plt.figure(figsize=(10, 5))
plt.plot(mean_tempo["decade"], mean_tempo["tempo"], linewidth=2)
plt.title('Tempo per Decade (BPM)')
plt.suptitle('')
plt.xlabel('Decade')
plt.ylabel('Tempo')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Check for trends in Duration column across different decades

```python
df["duration_s"] = df["duration_ms"] / 1000

df.groupby("decade").duration_s.agg(["min", "max", "mean"])
```

Boxplot for duration across different decades

```python
plt.figure(figsize=(12, 6))
df.boxplot(column='duration_s', by='decade')
plt.title('Duration (s) per Decade')
plt.suptitle('')
plt.xlabel('Decade')
plt.ylabel('Duration (s)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
average_duration = df.groupby('decade')['duration_s'].mean().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(average_duration['decade'], average_duration['duration_s'], marker='o', linewidth=2)
plt.title('Trend della durata media dei brani per decade')
plt.xlabel('Decade')
plt.ylabel('Durata media (secondi)')
plt.grid(True)
plt.xticks(average_duration['decade'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
max_duration = df.groupby("decade")["duration_s"].max().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(max_duration['decade'], max_duration['duration_s'], marker='o', linewidth=2)
plt.title('Trend of the max duration per decade')
plt.xlabel('Decade')
plt.ylabel('Max song duration (s)')
plt.grid(True)
plt.xticks(max_duration['decade'], rotation=45)
plt.tight_layout()
plt.show()
```

Analyze the trend of the Energy variable

```python
avg_energy_per_decade = df.groupby("decade").energy.mean().reset_index()

avg_energy_per_decade
```

```python
plt.figure(figsize=(10, 5))
plt.plot(avg_energy_per_decade['decade'], avg_energy_per_decade['energy'], linewidth=2)
plt.title('Trend of the average energy per decade')
plt.xlabel('Decade')
plt.ylabel('Energy')
plt.grid(True)
plt.xticks(avg_energy_per_decade['decade'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
avc_danceability_per_decade = df.groupby("decade").danceability.mean().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(avc_danceability_per_decade['decade'], avc_danceability_per_decade['danceability'], linewidth=2)
plt.title('Trend of the average danceability per decade')
plt.xlabel('Decade')
plt.ylabel('Energy')
plt.grid(True)
plt.xticks(avc_danceability_per_decade['decade'], rotation=45)
plt.tight_layout()
plt.show()
```

Correlation between Variables

```python
# Selezione delle sole colonne numeriche utili
feature_cols = [
    'danceability', 'acousticness', 'energy', 'instrumentalness', 'liveness',
    'loudness', 'speechiness', 'tempo', 'valence', 'duration_s', "decade"
]

# Creazione matrice di correlazione
corr_matrix = df[feature_cols].corr()

# Visualizzazione con heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
acousticness_per_decade = df.groupby("decade").acousticness.mean().reset_index()
acousticness_per_decade
```

```python
plt.figure(figsize=(10, 5))
plt.plot(acousticness_per_decade["decade"], acousticness_per_decade["acousticness"])
plt.title('Acousticness per decade')
plt.xlabel('Decade')
plt.ylabel('Acousticness')
plt.grid(True)
plt.xticks(acousticness_per_decade['decade'], rotation=45)
plt.tight_layout()
plt.show()
```
