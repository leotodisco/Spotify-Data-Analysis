Distribution Analysis

```python
import numpy as np
from scipy.stats import norm

valori_valence = df[df['decade'] == 2000]['valence']

mu = valori_valence.mean()
sigma = valori_valence.std()

x = np.linspace(0, 1, 1000)  
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(10, 5))
sns.histplot(valori_valence, bins=30, kde=True, stat='density', color='lightblue', edgecolor='black', label='Distribuzione reale')
plt.plot(x, y, color='red', linewidth=2, label=f'N({mu:.2f}, {sigma:.2f})')
plt.title('Valence - Anni 2000: Distribuzione reale vs Normale')
plt.xlabel('Valence')
plt.ylabel('Densità')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Dimensionality Reduction    

### Projection:


Prepare the data for the PCA Algorithm

```python
from sklearn.preprocessing import StandardScaler

numerical_labels = [
    'danceability', 'acousticness', 'energy', 'instrumentalness',
    'liveness', 'loudness', 'speechiness', 'tempo', 'valence', 'duration_s'
]

X = df[numerical_labels].dropna()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95) # preserves 95% of variance
X_pca = pca.fit_transform(X_scaled)

pca.n_components_
```

Clusteruing

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

scores = {}


for k in range(2, 8):
    kmeans = KMeans(n_clusters=k, random_state=42)
    y_pred = kmeans.fit_predict(X_pca)
    scores[k] = silhouette_score(X_pca, kmeans.labels_)
    
scores
```

```python
scores_df = pd.DataFrame({
    "k": scores.keys(),
    "score": scores.values()
})

plt.figure(figsize=(10, 5))
plt.plot(scores_df["k"], scores_df["score"], linewidth=2)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
y_pred = kmeans.fit_predict(X_pca)

```

```python

df_cluster = pd.DataFrame(X_pca[:, :2], columns=['PC1', 'PC2'])
df_cluster['Cluster'] = y_pred

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cluster, x='PC1', y='PC2', hue='Cluster', palette='tab10', s=60)
plt.title('Cluster KMeans su PCA - Visualizzazione 2D')
plt.grid(True)
plt.tight_layout()
plt.show()
```
